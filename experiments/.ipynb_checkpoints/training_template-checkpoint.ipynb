# StockAI Pro — Training Pipeline Template

**v2**: yfinance daily data · 11 technical indicators · RandomForest baseline  
Run `data_pipeline.py` first to populate `backend/app/cache/features/`.

## 0. Configuration

```python
import sys, os
from pathlib import Path

# Adjust to repo root so relative imports work when running from experiments/
REPO_ROOT = Path().resolve().parent
sys.path.insert(0, str(REPO_ROOT))

FEATURES_DIR = REPO_ROOT / 'backend' / 'app' / 'cache' / 'features'
MODELS_DIR   = REPO_ROOT / 'models'
MODELS_DIR.mkdir(parents=True, exist_ok=True)

TEST_SIZE    = 0.2    # 80/20 split
RANDOM_STATE = 42

print(f'Features dir : {FEATURES_DIR}')
print(f'Models dir   : {MODELS_DIR}')
csv_files = sorted(FEATURES_DIR.glob('*_features.csv'))
print(f'Feature CSVs found: {len(csv_files)}')
for f in csv_files:
    print(f'  {f.name}')
```

## 1. Load & Combine All Feature CSVs

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

frames = []
for csv_path in csv_files:
    ticker = csv_path.stem.replace('_features', '')
    df_t = pd.read_csv(csv_path, index_col='date', parse_dates=True)
    df_t['ticker'] = ticker
    frames.append(df_t)

if not frames:
    raise FileNotFoundError(
        f'No feature CSVs found in {FEATURES_DIR}.\n'
        'Run the pipeline first:\n'
        '  python -m backend.app.services.data_pipeline'
    )

df_all = pd.concat(frames, axis=0).sort_index()
df_all = df_all.dropna()

print(f'Combined dataset: {df_all.shape[0]:,} rows × {df_all.shape[1]} cols')
print(f'Date range      : {df_all.index.min().date()} → {df_all.index.max().date()}')
print(f'Tickers         : {df_all["ticker"].nunique()}')
print(f'\nTarget distribution:')
print(df_all['target'].value_counts(normalize=True).mul(100).round(2).astype(str) + '%')
df_all.head()
```

## 2. Feature Selection

```python
FEATURE_COLS = [
    # Price-derived
    'open', 'high', 'low', 'close', 'volume',
    # EMAs
    'ema9', 'ema15',
    # Momentum
    'rsi',
    # MACD
    'macd', 'macd_signal', 'macd_hist',
    # VWAP
    'vwap',
    # Trend
    'supertrend_dir', 'adx',
    # Volatility
    'bb_upper', 'bb_mid', 'bb_lower', 'bb_width', 'atr',
    # Volume
    'obv',
    # Stochastic
    'stoch_k', 'stoch_d',
    # Ichimoku
    'ich_tenkan', 'ich_kijun', 'ich_senkou_a', 'ich_senkou_b',
]

# Keep only columns available in this dataset
FEATURE_COLS = [c for c in FEATURE_COLS if c in df_all.columns]
TARGET_COL   = 'target'

X = df_all[FEATURE_COLS].values
y = df_all[TARGET_COL].values

print(f'Features used  : {len(FEATURE_COLS)}')
print(f'Feature columns: {FEATURE_COLS}')
print(f'X shape: {X.shape}, y shape: {y.shape}')
```

## 3. Train / Test Split (time-based)

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Time-based split (no shuffle — respects temporal order)
split_idx = int(len(X) * (1 - TEST_SIZE))
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

# Scale features
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)

print(f'Train: {X_train.shape[0]:,} rows  ({1-TEST_SIZE:.0%})')
print(f'Test : {X_test.shape[0]:,} rows  ({TEST_SIZE:.0%})')
```

## 4. RandomForest Baseline Model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    ConfusionMatrixDisplay, roc_auc_score
)
import matplotlib.pyplot as plt

print('Training RandomForest …')
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=8,
    min_samples_leaf=20,
    n_jobs=-1,
    random_state=RANDOM_STATE,
    class_weight='balanced'
)
rf.fit(X_train_sc, y_train)
print('Done.')
```

## 5. Evaluation — Accuracy & Confusion Matrix

```python
y_pred = rf.predict(X_test_sc)
y_prob = rf.predict_proba(X_test_sc)[:, 1]

acc    = accuracy_score(y_test, y_pred)
auc    = roc_auc_score(y_test, y_prob)

print('=' * 50)
print(f'  Accuracy  : {acc:.4f}  ({acc*100:.2f}%)')
print(f'  ROC-AUC   : {auc:.4f}')
print('=' * 50)
print()
print(classification_report(y_test, y_pred, target_names=['DOWN(0)', 'UP(1)']))

# Plot confusion matrix
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['DOWN', 'UP'])
disp.plot(ax=axes[0], colorbar=False, cmap='Blues')
axes[0].set_title('Confusion Matrix')

# Feature importance
importances = pd.Series(rf.feature_importances_, index=FEATURE_COLS).sort_values(ascending=True)
importances.tail(15).plot(kind='barh', ax=axes[1], color='steelblue')
axes[1].set_title('Top 15 Feature Importances')
axes[1].set_xlabel('Importance')

plt.tight_layout()
plt.savefig(str(MODELS_DIR / 'rf_baseline_eval.png'), dpi=150, bbox_inches='tight')
plt.show()
print(f'Plot saved to: {MODELS_DIR / "rf_baseline_eval.png"}')
```

## 6. Save Model Artifact

```python
import joblib
from datetime import date

model_dir = MODELS_DIR / f'rf_baseline_{date.today().strftime("%Y-%m-%d")}'
model_dir.mkdir(parents=True, exist_ok=True)

joblib.dump(rf,     model_dir / 'rf_model.pkl')
joblib.dump(scaler, model_dir / 'scaler.pkl')

# Save feature list for inference
import json
with open(model_dir / 'feature_cols.json', 'w') as f:
    json.dump(FEATURE_COLS, f, indent=2)

print(f'Model saved to: {model_dir}')
print(f'Files: rf_model.pkl, scaler.pkl, feature_cols.json')
```

## 7. (Optional) XGBoost Comparison

```python
try:
    import xgboost as xgb
    from sklearn.metrics import accuracy_score

    xgb_clf = xgb.XGBClassifier(
        n_estimators=300,
        max_depth=6,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        use_label_encoder=False,
        eval_metric='logloss',
        random_state=RANDOM_STATE,
        n_jobs=-1,
    )
    xgb_clf.fit(
        X_train_sc, y_train,
        eval_set=[(X_test_sc, y_test)],
        verbose=50,
    )
    xgb_acc = accuracy_score(y_test, xgb_clf.predict(X_test_sc))
    print(f'XGBoost Accuracy: {xgb_acc:.4f}  ({xgb_acc*100:.2f}%)')
    joblib.dump(xgb_clf, model_dir / 'xgb_model.pkl')
    print(f'XGBoost model saved to: {model_dir / "xgb_model.pkl"}')
except ImportError:
    print('xgboost not installed – skipping. pip install xgboost')
```
